```
from PIL import Image, ImageDraw, ImageFont
import os

def create_icon(size, filename):
    # Create a new image with a solid blue background
    # Mode 'RGBA' ensures we have an alpha channel (even if we don't use transparency to be safe)
    img = Image.new('RGBA', (size, size), color=(59, 130, 246)) # Tailwind Blue-500

    draw = ImageDraw.Draw(img)

    # We can't easily load custom fonts in this environment without external files,
    # so we'll draw a simple white rectangle in the center to represent "content"
    # or just keep it solid blue. Let's make it a "Test" icon.

    # Draw a white circle in the middle
    padding = size // 4
    draw.ellipse([padding, padding, size - padding, size - padding], fill=(255, 255, 255))

    # Ensure directory exists
    if not os.path.exists('icons'):
        os.makedirs('icons')

    img.save(f'icons/{filename}')
    print(f"Generated {filename}")

# Chrome Extensions typically need these 4 sizes
sizes = {
    16: 'icon16.png',
    32: 'icon32.png',
    48: 'icon48.png',
    128: 'icon128.png'
}

print("Generating extension icons...")
for size, name in sizes.items():
    create_icon(size, name)

print("Done! Icons are in the 'icons' folder.")
```
